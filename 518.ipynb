```python
import re
import requests
from bs4 import BeautifulSoup
import pandas as pd

resp = requests.get("http://yp.518.com.tw/service-life.html?ctf=10")
soup = BeautifulSoup(resp.text, "lxml")
soup
```

```python
name_phone = [tag.text for tag in soup.select("li.comp_tel")]
name_phone
```

```python
name = [x.split("/")[0].strip() for x in name_phone]
name
```

```python
name_phone_str = "".join(name_phone)
name_phone_str
```

```python
phone = re.findall("[0-9]{2}-[0-9]+", name_phone_str)
phone
```

```python
address = [x.text.replace('/','').replace(' ','') for x in soup.select("li.comp_loca")]
address
```

```python
df = pd.DataFrame({"店名":name,
                   "地址": address,
                   "電話": phone},columns = ["店名","地址","電話"])
df
```

```python
df.to_csv("csv_results.csv", index =False, encoding="utf-8")
```
